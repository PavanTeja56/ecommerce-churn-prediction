```python
import pandas as pd
import json

pd.set_option("display.max_columns",None)
df_clean = pd.read_csv(r"C:\Users\PAVAN TEJA\Downloads\projects\ecommerce-churn-prediction\data\processed\cleaned_transactions.csv",encoding="latin1")
print(f"Cleand data set size{df_clean.shape}")
df_clean.head()
```

```python
assert df_clean.isnull().sum().sum()==0, "Missing values found!"
print("CHECK 1 PASSED : No missing values")
```

```python
assert (df_clean["Quantity"]>0).all(), "Negative quantities found!"
print("CHECK 2 PASSED : All quantites are positive")
```

```python
assert (df_clean["UnitPrice"]>0).all(), "Invalied prices found"
print("CHECK 3 PASSED : All prices are positive")
```

```python
assert df_clean["CustomerID"].dtype=="int64", "Customer ID is not integer!"
print("CHECK 4 PASSED : Customer ID is integer type")
```

```python
df_clean["InvoiceDate"]=pd.to_datetime(df_clean["InvoiceDate"])
print("Date Range:")
print("Start:",df_clean["InvoiceDate"].min())
print("End  :",df_clean["InvoiceDate"].max())
```

```python
summary_stats = {
    "total_rows": int(len(df_clean)),
    "total_columns": int(len(df_clean.columns)),
    "unique_customers": int(df_clean["CustomerID"].nunique()),
    "unique_products": int(df_clean["StockCode"].nunique()),
    "unique_countries": int(df_clean["Country"].nunique()),
    "total_revenue": round(float(df_clean["TotalPrice"].sum()), 2),
    "average_order_value": round(
        df_clean.groupby("InvoiceNo")["TotalPrice"].sum().mean(), 2
    )
}

summary_stats
```

```python
validation_report = {
    "total_rows": summary_stats["total_rows"],
    "total_columns": summary_stats["total_columns"],
    "date_range": {
        "start": str(df_clean["InvoiceDate"].min()),
        "end": str(df_clean["InvoiceDate"].max())
    },
    "unique_customers": summary_stats["unique_customers"],
    "unique_products": summary_stats["unique_products"],
    "unique_countries": summary_stats["unique_countries"],
    "total_revenue": summary_stats["total_revenue"],
    "average_order_value": summary_stats["average_order_value"],
    "validation_passed": True,
    "checks": {
        "no_missing_values": True,
        "all_quantities_positive": True,
        "all_prices_positive": True,
        "customer_id_is_integer": True
    }
}

with open(r"C:\Users\PAVAN TEJA\Downloads\projects\ecommerce-churn-prediction\data\processed\validation_report.json", "w") as f:
    json.dump(validation_report, f, indent=4)

print("‚úÖ Validation report saved to data/processed/validation_report.json")
```
